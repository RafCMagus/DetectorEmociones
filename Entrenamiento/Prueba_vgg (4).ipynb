```python
import os
import csv
import numpy as np
import pandas as pd
import h5py
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
from torchvision import transforms
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score, recall_score
from sklearn.model_selection import KFold
from tqdm import tqdm
import matplotlib.pyplot as plt
from torch.optim.lr_scheduler import ReduceLROnPlateau
from torch.utils.tensorboard import SummaryWriter
from PIL import Image
import shutil


# Cargar Google Drive y Preprocesar el dataset FER2013
from google.colab import drive
drive.mount('/content/drive')

# Cargar el CSV de FER2013
csv_path = '/content/drive/MyDrive/dataset3/fer2013.csv'
df = pd.read_csv(csv_path)

# Crear carpetas por clase y convertir datos numéricos a imágenes
output_dir = '/content/FER_images'
os.makedirs(output_dir, exist_ok=True)

# Crear directorio del modelo
os.makedirs('/content/model_checkpoints', exist_ok=True)

emotion_labels = {0: 'Angry', 1: 'Disgust', 2: 'Fear', 3: 'Happy', 4: 'Sad', 5: 'Surprise', 6: 'Neutral'}

for label in emotion_labels.values():
    os.makedirs(os.path.join(output_dir, label), exist_ok=True)

for index, row in tqdm(df.iterrows(), total=df.shape[0]):
    emotion = emotion_labels[row['emotion']]
    pixels = np.fromstring(row['pixels'], sep=' ', dtype=int)
    image = pixels.reshape(48, 48)
    img = Image.fromarray(image.astype('uint8'), mode='L')
    img_path = os.path.join(output_dir, emotion, f'{index}.png')
    img.save(img_path)

# Importar librerías necesarias
import os

# Directorio donde están organizadas las imágenes
output_dir = '/content/FER_images'

# Diccionario para almacenar el conteo por clase
image_counts = {}
for label, emotion in emotion_labels.items():
    class_dir = os.path.join(output_dir, emotion)
    image_counts[label] = len(os.listdir(class_dir))

# Mostrar la distribución actual de imágenes por clase
print("Distribución de imágenes por clase:")
for label, count in image_counts.items():
    print(f"Clase {emotion_labels[label]} ({label}): {count} imágenes")

# Visualizar la distribución de clases
import matplotlib.pyplot as plt

labels = [emotion_labels[label] for label in image_counts.keys()]
counts = list(image_counts.values())

plt.bar(labels, counts, alpha=0.7)
plt.title("Distribución de imágenes por clase")
plt.xlabel("Clases (Carpetas)")
plt.ylabel("Número de imágenes")
plt.xticks(rotation=45)
plt.show()


import os
import random
from PIL import Image
import matplotlib.pyplot as plt
from torchvision import transforms

# Directorio donde están organizadas las imágenes
output_dir = '/content/FER_images'
emotion_labels = {0: 'Angry', 1: 'Disgust', 2: 'Fear', 3: 'Happy', 4: 'Sad', 5: 'Surprise', 6: 'Neutral'}

# Recalcular image_counts para cada clase
image_counts = {}
for label, emotion in emotion_labels.items():
    class_dir = os.path.join(output_dir, emotion)
    image_counts[label] = len(os.listdir(class_dir))

print("Distribución de imágenes por clase antes del balanceo:")
for label, count in image_counts.items():
    print(f"Clase {emotion_labels[label]} ({label}): {count} imágenes")

# Nivel objetivo basado en la clase "Neutral"
target_level = image_counts[6]  # Neutral

# Reducir "Happy" si tiene más imágenes que el nivel objetivo
random.seed(42)  # Para garantizar reproducibilidad
happy_dir = os.path.join(output_dir, "Happy")
happy_images = os.listdir(happy_dir)
if len(happy_images) > target_level:
    happy_to_keep = random.sample(happy_images, target_level)
    for img_name in happy_images:
        if img_name not in happy_to_keep:
            os.remove(os.path.join(happy_dir, img_name))
    print(f"Reducida la clase 'Happy' a {target_level} imágenes.")

# Recalcular image_counts después de reducir "Happy"
image_counts = {}
for label, emotion in emotion_labels.items():
    class_dir = os.path.join(output_dir, emotion)
    image_counts[label] = len(os.listdir(class_dir))

# Calcular cuántas imágenes adicionales necesita cada clase
adjusted_augmentation = {}
for label, count in image_counts.items():
    if count < target_level:
        adjusted_augmentation[label] = target_level - count

print("\nAumentos necesarios para balancear las clases:")
for label, extra in adjusted_augmentation.items():
    print(f"Clase {emotion_labels[label]} ({label}): {extra} imágenes adicionales")

# Función para generar imágenes aumentadas

def generate_augmented_images(class_dir, augment_count, transform):
    image_files = os.listdir(class_dir)
    random.seed(42)  # Para reproducibilidad
    for i in range(augment_count):
        img_name = random.choice(image_files)
        img_path = os.path.join(class_dir, img_name)
        image = Image.open(img_path).convert('RGB')
        augmented_image = transform(image)

        # Guardar la imagen aumentada
        augmented_path = os.path.join(class_dir, f"augmented_{i}_{img_name}")
        augmented_image.save(augmented_path)

# Transformaciones específicas para las clases
targeted_augmentation = {
    0: transforms.Compose([
        transforms.RandomRotation(30),
        transforms.ColorJitter(brightness=0.4, contrast=0.4),
        transforms.Resize((48, 48))
    ]),
    1: transforms.Compose([
        transforms.RandomHorizontalFlip(),
        transforms.ColorJitter(hue=0.5),
        transforms.Resize((48, 48))
    ]),
    2: transforms.Compose([
        transforms.RandomRotation(40),
        transforms.ColorJitter(brightness=0.5, contrast=0.5),
        transforms.Resize((48, 48))
    ]),
    4: transforms.Compose([
        transforms.RandomRotation(25),
        transforms.RandomHorizontalFlip(),
        transforms.ColorJitter(saturation=0.3),
        transforms.Resize((48, 48))
    ]),
    5: transforms.Compose([
        transforms.ColorJitter(saturation=0.7),
        transforms.RandomResizedCrop(size=44, scale=(0.7, 1.0)),
        transforms.Resize((48, 48))
    ])
}

# Generar imágenes aumentadas para balancear las clases
for label, augment_count in adjusted_augmentation.items():
    class_dir = os.path.join(output_dir, emotion_labels[label])
    transform = targeted_augmentation.get(label, None)
    if transform:
        generate_augmented_images(class_dir, augment_count, transform)

# Recalcular image_counts después del aumento
image_counts = {}
for label, emotion in emotion_labels.items():
    class_dir = os.path.join(output_dir, emotion)
    image_counts[label] = len(os.listdir(class_dir))

print("\nDistribución de imágenes por clase después del balanceo:")
for label, count in image_counts.items():
    print(f"Clase {emotion_labels[label]} ({label}): {count} imágenes")

# Verificar si las clases están balanceadas
def verificar_balance(image_counts, target_level):
    for label, count in image_counts.items():
        if count != target_level:
            print(f"Clase {emotion_labels[label]} ({label}) no está balanceada: {count} imágenes (objetivo: {target_level})")
        else:
            print(f"Clase {emotion_labels[label]} ({label}) está correctamente balanceada: {count} imágenes")

verificar_balance(image_counts, target_level)

# Visualizar la distribución ajustada
labels = [emotion_labels[label] for label in image_counts.keys()]
counts = list(image_counts.values())
plt.bar(labels, counts, alpha=0.7)
plt.title("Distribución ajustada de imágenes por clase")
plt.xlabel("Clases (Carpetas)")
plt.ylabel("Número de imágenes")
plt.xticks(rotation=45)
plt.show()


# Dataset personalizado con transformaciones actualizadas
class FER2013Dataset(Dataset):
    def __init__(self, root_dir, transform=None, targeted_augmentation=None, augment_counts=None):
        self.image_paths = []
        self.labels = []
        self.transform = transform
        self.targeted_augmentation = targeted_augmentation
        self.augment_counts = augment_counts or {}

        for label, emotion in emotion_labels.items():
            emotion_dir = os.path.join(root_dir, emotion)
            image_files = os.listdir(emotion_dir)

            # Procesar imágenes originales
            for img_name in image_files:
                img_path = os.path.join(emotion_dir, img_name)
                self.image_paths.append(img_path)
                self.labels.append(label)

            # Generar imágenes aumentadas si es necesario
            if label in self.augment_counts and self.targeted_augmentation:
                transform = self.targeted_augmentation.get(label, None)
                if transform:
                    for img_name in image_files:
                        img_path = os.path.join(emotion_dir, img_name)
                        image = Image.open(img_path).convert('RGB')
                        for _ in range(self.augment_counts[label] // len(image_files)):
                            augmented_image = transform(image)
                            self.image_paths.append(img_path)  # Usar la ruta original
                            self.labels.append(label)

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        # Verificar si el elemento es una ruta de imagen o una imagen aumentada
        image_path_or_image = self.image_paths[idx]
        label = self.labels[idx]

        image = Image.open(image_path_or_image).convert('RGB')

        if self.transform:
            image = self.transform(image)
        return image, label

# Transformaciones generales para las imágenes de entrenamiento
transform_train_general = transforms.Compose([
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(degrees=20),  # Incrementar rango de rotación
    #transforms.RandomAffine(degrees=20, translate=(0.1, 0.1)),  # Agregar deformaciones
    transforms.ColorJitter(brightness=0.3, contrast=0.3),
    #transforms.RandomResizedCrop(size=48, scale=(0.7, 1.0)),  # Incrementar recortes aleatorios
    #transforms.RandomErasing(p=0.5, scale=(0.02, 0.2)),
    transforms.Resize((48, 48)),
    transforms.ToTensor(),
    transforms.Normalize([0.5], [0.5])  # Normalización
])

# Transformaciones para las imágenes de prueba
transform_test = transforms.Compose([
    transforms.CenterCrop(48),  # Recorte centrado para mantener consistencia
    transforms.ToTensor(),
    transforms.Normalize([0.5], [0.5])
])

# Cargar el dataset con las nuevas transformaciones
dataset_train = FER2013Dataset(
    root_dir=output_dir,
    transform=transform_train_general
)
dataset_test = FER2013Dataset(
    root_dir=output_dir,
    transform=transform_test
)

# Balanceo de clases para el DataLoader
#class_counts = np.bincount(dataset_train.labels)
#class_weights = 1.0 / class_counts
#sample_weights = [class_weights[label] for label in dataset_train.labels]
#sampler = WeightedRandomSampler(sample_weights, len(sample_weights))

# Crear DataLoaders
train_loader = DataLoader(dataset_train, batch_size=128, shuffle=True)
test_loader = DataLoader(dataset_test, batch_size=128, shuffle=False)

# Visualizar la distribución final en el DataLoader
class_counts_final = np.bincount(dataset_train.labels)
plt.bar([emotion_labels[i] for i in range(len(class_counts_final))], class_counts_final, alpha=0.7)
plt.title("Distribución final de imágenes por clase en el DataLoader")
plt.xlabel("Clases (Carpetas)")
plt.ylabel("Número de imágenes")
plt.xticks(rotation=45)
plt.show()





import matplotlib.pyplot as plt

def visualizar_aumentos(clase, num_muestras=5):
    """Visualizar imágenes originales y sus versiones aumentadas."""
    clase_nombre = emotion_labels[clase]
    clase_dir = os.path.join(output_dir, clase_nombre)
    imagenes = os.listdir(clase_dir)[:num_muestras]  # Tomar las primeras imágenes de la clase

    fig, axes = plt.subplots(num_muestras, 2, figsize=(8, num_muestras * 4))

    for i, img_name in enumerate(imagenes):
        img_path = os.path.join(clase_dir, img_name)
        imagen_original = Image.open(img_path).convert('RGB')

        # Aplicar aumentos dirigidos
        if clase in targeted_augmentation:
            transform = targeted_augmentation[clase]
            imagen_aumentada = transform(imagen_original)
        else:
            imagen_aumentada = imagen_original

        # Mostrar imagen original
        axes[i, 0].imshow(imagen_original)
        axes[i, 0].set_title(f"Original: {clase_nombre}")
        axes[i, 0].axis('off')

        # Mostrar imagen aumentada
        axes[i, 1].imshow(imagen_aumentada)
        axes[i, 1].set_title(f"Aumentada: {clase_nombre}")
        axes[i, 1].axis('off')

    plt.tight_layout()
    plt.show()

# Ejemplo: Visualizar aumentos para la clase "Disgust" (1)
visualizar_aumentos(clase=1)


from PIL import ImageStat

def calcular_estadisticas(imagen):
    """Calcular estadísticas básicas de una imagen."""
    stat = ImageStat.Stat(imagen)
    brillo = stat.mean[0]  # Promedio de brillo
    contraste = stat.stddev[0]  # Desviación estándar del brillo
    return brillo, contraste

def analizar_estadisticas(clase, num_muestras=10):
    """Comparar estadísticas entre imágenes originales y aumentadas."""
    clase_nombre = emotion_labels[clase]
    clase_dir = os.path.join(output_dir, clase_nombre)
    imagenes = os.listdir(clase_dir)[:num_muestras]

    estadisticas_original = []
    estadisticas_aumentadas = []

    for img_name in imagenes:
        img_path = os.path.join(clase_dir, img_name)
        imagen_original = Image.open(img_path).convert('RGB')

        # Estadísticas de la imagen original
        estadisticas_original.append(calcular_estadisticas(imagen_original))

        # Estadísticas de la imagen aumentada
        if clase in targeted_augmentation:
            transform = targeted_augmentation[clase]
            imagen_aumentada = transform(imagen_original)
            estadisticas_aumentadas.append(calcular_estadisticas(imagen_aumentada))

    # Promedio de estadísticas
    promedio_original = np.mean(estadisticas_original, axis=0)
    promedio_aumentado = np.mean(estadisticas_aumentadas, axis=0)

    print(f"Clase: {clase_nombre}")
    print(f"Brillo original: {promedio_original[0]:.2f}, Contraste original: {promedio_original[1]:.2f}")
    print(f"Brillo aumentado: {promedio_aumentado[0]:.2f}, Contraste aumentado: {promedio_aumentado[1]:.2f}")

# Ejemplo: Analizar estadísticas para la clase "Disgust" (1)
analizar_estadisticas(clase=1)





# Modelo VGG Personalizado
class VGG(nn.Module):
    def __init__(self, num_classes=7):
        super(VGG, self).__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(128, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.Conv2d(256, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(256, 512, kernel_size=3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(inplace=True),
            nn.Conv2d(512, 512, kernel_size=3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(512, 512, kernel_size=3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(inplace=True),
            nn.Conv2d(512, 512, kernel_size=3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )
        self.classifier = nn.Sequential(
            nn.Linear(512 * 1 * 1, 4096),
            nn.ReLU(True),
            nn.Dropout(0.8),
            nn.Linear(4096, 4096),
            nn.ReLU(True),
            nn.Dropout(0.6),
            nn.Linear(4096, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        x = x.view(x.size(0), -1)
        x = self.classifier(x)
        return x

# Configuración de entrenamiento
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = VGG().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.01, momentum=0.9, weight_decay=1e-3)
scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.5, patience=5, verbose=True)

# TensorBoard
writer = SummaryWriter()

# Entrenamiento con k-fold y Early Stopping
def train_model_kfold(model, dataset_train, criterion, optimizer, scheduler, k_folds=5, num_epochs=50):
    kfold = KFold(n_splits=k_folds, shuffle=True)
    os.makedirs('/content/model_checkpoints', exist_ok=True)

    for fold, (train_idx, val_idx) in enumerate(kfold.split(range(len(dataset_train)))):
        print(f'Fold {fold + 1}/{k_folds}')

        # Subsets de entrenamiento y validación
        train_subset = torch.utils.data.Subset(dataset_train, train_idx)
        val_subset = torch.utils.data.Subset(dataset_train, val_idx)

        train_loader = DataLoader(train_subset, batch_size=128, shuffle=True)  # Eliminamos el sampler
        val_loader = DataLoader(val_subset, batch_size=128, shuffle=False)

        best_acc = 0.0  # Reset para cada fold
        for epoch in range(num_epochs):
            # Entrenamiento
            model.train()
            running_loss, correct, total = 0.0, 0, 0
            for inputs, labels in tqdm(train_loader):
                inputs, labels = inputs.to(device), labels.to(device)
                optimizer.zero_grad()
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()

                running_loss += loss.item() * inputs.size(0)
                _, preds = outputs.max(1)
                correct += preds.eq(labels).sum().item()
                total += labels.size(0)

            train_acc = 100. * correct / total

            # Validación
            model.eval()
            val_loss, val_correct, val_total = 0.0, 0, 0
            with torch.no_grad():
                for inputs, labels in val_loader:
                    inputs, labels = inputs.to(device), labels.to(device)
                    outputs = model(inputs)
                    loss = criterion(outputs, labels)
                    val_loss += loss.item() * inputs.size(0)
                    _, preds = outputs.max(1)
                    val_correct += preds.eq(labels).sum().item()
                    val_total += labels.size(0)

            val_acc = 100. * val_correct / val_total
            scheduler.step(val_acc)

            print(f"Fold {fold + 1} Epoch {epoch + 1}/{num_epochs}, Train Acc: {train_acc:.2f}%, Val Acc: {val_acc:.2f}%")

            # Guardar el mejor modelo del fold actual
            if val_acc > best_acc:
                best_acc = val_acc
                model_path = f'/content/model_checkpoints/best_model_fold_{fold + 1}.pth'
                torch.save(model.state_dict(), model_path)

    print("Modelos de los folds guardados correctamente.")

    # Descargar los modelos de cada fold
    print("\nDescargando los modelos generados de cada fold...")
    from google.colab import files
    for fold in range(k_folds):
        model_path = f'/content/model_checkpoints/best_model_fold_{fold + 1}.pth'
        if os.path.exists(model_path):
            files.download(model_path)



# Evaluación
# Evaluación de los modelos por fold
def evaluate_models_by_fold(model_class, test_loader, model_dir, k_folds=5):
    for fold in range(k_folds):
        model_path = os.path.join(model_dir, f'best_model_fold_{fold + 1}.pth')

        if os.path.exists(model_path):
            print(f"\nEvaluando el modelo del Fold {fold + 1}...")

            # Cargar el modelo
            model = model_class().to(device)
            model.load_state_dict(torch.load(model_path))
            model.eval()

            preds, labels = [], []
            with torch.no_grad():
                for inputs, targets in test_loader:
                    inputs, targets = inputs.to(device), targets.to(device)
                    outputs = model(inputs)
                    _, predicted = outputs.max(1)
                    preds.extend(predicted.cpu().numpy())
                    labels.extend(targets.cpu().numpy())

            # Métricas
            acc = accuracy_score(labels, preds)
            f1 = f1_score(labels, preds, average='weighted')
            recall = recall_score(labels, preds, average='weighted')
            conf_matrix = confusion_matrix(labels, preds)

            print("Classification Report:")
            print(classification_report(labels, preds))
            print(f"Accuracy: {acc:.4f}, F1: {f1:.4f}, Recall: {recall:.4f}")

            # Mostrar matriz de confusión
            plt.figure(figsize=(10, 8))
            plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.Blues)
            plt.colorbar()
            for i in range(conf_matrix.shape[0]):
                for j in range(conf_matrix.shape[1]):
                    plt.text(j, i, conf_matrix[i, j], horizontalalignment="center", color="red")
            plt.title(f"Confusion Matrix for Fold {fold + 1}")
            plt.ylabel("True Label")
            plt.xlabel("Predicted Label")
            plt.show()
        else:
            print(f"El modelo para el Fold {fold + 1} no fue encontrado en {model_path}.")



# Ejecutar entrenamiento y evaluación
train_model_kfold(model, dataset_train, criterion, optimizer, scheduler, k_folds=5, num_epochs=50)
# Llamar a la evaluación
evaluate_models_by_fold(VGG, test_loader, model_dir='/content/model_checkpoints', k_folds=5)
```

```python
import torch
import numpy as np
from sklearn.metrics import classification_report, accuracy_score, f1_score, recall_score, confusion_matrix
import matplotlib.pyplot as plt

# Crear el dataset de probabilidades para el conjunto de prueba
def generate_test_meta_dataset(test_loader, base_models, device):
    meta_features = []
    meta_labels = []

    for images, labels in test_loader:
        images = images.to(device)
        labels = labels.to(device)

        # Generar probabilidades para cada modelo base
        model_outputs = []
        for model in base_models:
            model.eval()
            with torch.no_grad():
                outputs = model(images)
                probabilities = torch.softmax(outputs, dim=1)
                model_outputs.append(probabilities.cpu().numpy())

        # Concatenar las probabilidades de todos los modelos base
        combined_outputs = np.concatenate(model_outputs, axis=1)
        meta_features.append(combined_outputs)
        meta_labels.extend(labels.cpu().numpy())

    # Convertir las listas a arrays de NumPy
    meta_features = np.vstack(meta_features)
    meta_labels = np.array(meta_labels)

    return meta_features, meta_labels

# Crear el DataLoader para el conjunto de prueba del metamodelo
def create_test_loader_meta(test_loader, base_models, device):
    test_meta_features, test_meta_labels = generate_test_meta_dataset(test_loader, base_models, device)
    test_dataset_meta = torch.utils.data.TensorDataset(
        torch.tensor(test_meta_features, dtype=torch.float32),
        torch.tensor(test_meta_labels, dtype=torch.long)
    )
    return torch.utils.data.DataLoader(test_dataset_meta, batch_size=128, shuffle=False)

# Evaluar el metamodelo
def evaluate_meta_model(meta_model, test_loader_meta, device):
    meta_model.eval()
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for features, labels in test_loader_meta:
            features = features.to(device)
            labels = labels.to(device)
            outputs = meta_model(features)
            _, preds = outputs.max(1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    # Calcular métricas de evaluación
    acc = accuracy_score(all_labels, all_preds)
    f1 = f1_score(all_labels, all_preds, average='weighted')
    recall = recall_score(all_labels, all_preds, average='weighted')
    conf_matrix = confusion_matrix(all_labels, all_preds)

    print("Classification Report:")
    print(classification_report(all_labels, all_preds))
    print(f"Accuracy: {acc:.4f}, F1: {f1:.4f}, Recall: {recall:.4f}")

    # Mostrar matriz de confusión
    plt.figure(figsize=(10, 8))
    plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.Blues)
    plt.colorbar()
    for i in range(conf_matrix.shape[0]):
        for j in range(conf_matrix.shape[1]):
            plt.text(j, i, conf_matrix[i, j], horizontalalignment="center", color="red")
    plt.title("Confusion Matrix")
    plt.ylabel("True Label")
    plt.xlabel("Predicted Label")
    plt.show()

# Generar el DataLoader para el metamodelo
test_loader_meta = create_test_loader_meta(test_loader, models, device)

# Evaluar el metamodelo
evaluate_meta_model(meta_model, test_loader_meta, device)
```

```python
import cv2
import dlib
import numpy as np
from PIL import Image
import torch
from torchvision import transforms
from google.colab import files

# Cargar el detector de rostros de dlib
detector = dlib.get_frontal_face_detector()

# Mapeo de etiquetas de emociones
emotion_label_map = {
    0: 'Angry',
    1: 'Disgust',
    2: 'Fear',
    3: 'Happy',
    4: 'Sad',
    5: 'Surprise',
    6: 'Neutral'
}

# Ruta exacta al archivo del metamodelo
meta_model_path = '/content/meta_model.pth'

# Asegúrate de haber definido la clase MetaModel antes de cargar
input_size = len(models) * 7  # Número de modelos base * número de clases
meta_model = MetaModel(input_size=input_size, num_classes=7).to(device)

# Cargar los pesos del metamodelo desde la ruta
meta_model.load_state_dict(torch.load(meta_model_path, map_location=device))
meta_model.eval()  # Configurar el modelo en modo evaluación

# Transformación para las imágenes (sin Grayscale para mantener los 3 canales)
image_transform = transforms.Compose([
    transforms.Resize((48, 48)),  # Redimensionar la imagen al tamaño esperado
    transforms.ToTensor(),       # Convertir a tensor
    transforms.Normalize([0.5, 0.5, 0.5], [0.5, 0.5, 0.5])  # Normalización
])

# Función para predecir las emociones en una imagen
def predict_emotions(image_path, meta_model, models):
    # Leer la imagen
    image = cv2.imread(image_path)
    rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convertir a RGB para el modelo

    # Detectar rostros
    faces = detector(rgb_image)
    print(f"{len(faces)} rostros detectados")

    # Verificar si hay rostros
    if len(faces) == 0:
        print("No se detectaron rostros en la imagen.")
        return

    # Procesar cada rostro
    for i, face in enumerate(faces):
        # Recortar el rostro detectado
        x, y, w, h = face.left(), face.top(), face.width(), face.height()
        face_img = rgb_image[y:y+h, x:x+w]

        # Convertir el rostro recortado al formato esperado
        face_pil = Image.fromarray(face_img)
        face_tensor = image_transform(face_pil).unsqueeze(0).to(device)

        # Obtener las probabilidades de los modelos base
        probs = []
        for model in models:
            model.eval()
            with torch.no_grad():
                output = model(face_tensor)
                probs.append(torch.softmax(output, dim=1).cpu().numpy())

        # Usar el metamodelo para predecir
        probs_stack = np.hstack(probs)
        probs_tensor = torch.tensor(probs_stack, dtype=torch.float32).to(device)
        meta_model.eval()
        with torch.no_grad():
            meta_output = meta_model(probs_tensor)
            meta_pred = torch.argmax(meta_output, dim=1).item()

        # Verificar si la predicción está en el mapeo
        if meta_pred not in emotion_label_map:
            print(f"Error: Predicción {meta_pred} no válida.")
            continue

        # Mostrar la predicción
        emotion_label = emotion_label_map[meta_pred]
        print(f"Rostro {i+1}: {emotion_label}")

        # Dibujar el rostro y la emoción en la imagen original
        cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)
        cv2.putText(image, emotion_label, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

    # Guardar la imagen procesada en Colab y mostrarla
    output_path = "/content/processed_image.jpg"
    cv2.imwrite(output_path, image)
    print(f"Resultados guardados en: {output_path}")
    from google.colab.patches import cv2_imshow
    cv2_imshow(image)

# Subir una imagen desde tu PC
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# Llamar a la función con tu metamodelo y modelos base
predict_emotions(image_path, meta_model, models)
```
